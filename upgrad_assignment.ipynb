```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

Loading the data

```python
data = pd.read_csv('loan.csv',index_col=0)
data.head()
```

```python
data.shape
```

Checking the missing values

```python
data.isnull().sum()
```

Finding columns which contain more than 50% missing data

```python
check = data.isnull().sum() / len(data) 
check
```

Deleting columns with 50% null values

```python
cols = check[check < 0.5].index
data = data[cols]
data
```

```python
data.shape
```

```python
data.info(verbose=True, show_counts=True)
```

Some columns are of type object which should be of type int or float. Lets check such columns

```python
#checking for int_rate
data.int_rate.value_counts()
```

```python
#we notice the values end with % so we can trim the extra % sign to change the type to float
data['int_rate'] = data['int_rate'].apply(lambda x: float(x[:-1]))
```

```python
#checking for revol_util
data.revol_util.value_counts()
```

```python
data.revol_util.isnull().sum()
```

```python
def convertToFloat(x):
  try:
    if(str(x) == "nan" or x == None or str(x)== "NaN"):
      x = 0
    else:
      x = float(x[:-1])
  except:
    print("error converting " + str(x))
  return x
```

```python
#we notice the values end with % so we can trim the extra % sign to change the type to float
data['revol_util'] = data['revol_util'].apply(lambda x: convertToFloat(x))
```

```python
data.corr()
```

```python
numeric_var = [key for key in dict(data.dtypes)
                   if dict(data.dtypes)[key]
                       in ['float64','float32','int32','int64']] # Numeric Variable

cat_var = [key for key in dict(data.dtypes)
             if dict(data.dtypes)[key] in ['object'] ]
```

```python
for x in cat_var:
  print('=======')
  print(x)
  print(data[x].describe())
```

```python
len(numeric_var) + len(cat_var) == data.shape[1]
```

```python
#from the description we see that emp_title, last_credit_pull_d, last_pymnt_d, earliest_cr_line, title, desc, emp_title have too many unique values 
# removing it from the list of categorical columns. We will do analysis on these separately

additional_cols = ['emp_title', 'emp_length', 'issue_d', 'last_credit_pull_d', "last_pymnt_d", "earliest_cr_line", "title", "desc"]
for c in additional_cols:
  cat_var.remove(c)
```

```python
#we can see that application_type, initial_list_status, pymnt_plan have onnly one unique value, they wont contribute anything in the analysis as the values are same everywhere
# we can drop these columns

for c in ['application_type', 'initial_list_status', 'pymnt_plan']:
  cat_var.remove(c)
```

```python
#we also see that the value of url is unique for each row so this is also not very useful to draw an analysis
#dropping that
cat_var.remove('url')
```

```python
#the zip code has 5 digits and last two digits are masked. The first digit in US zip code is for a state and we have another column for that.
#we can remove the last two digits and first digit so we can see if we can draw any analysis from 
#the region which the person belongs to which are identified by 2nd and 3rd digit in the zip code

data['zip_code'] = data['zip_code'].apply(lambda x: x[1:3])
```

```python
data['zip_code'].describe()
```

```python
#we now see there are only 100 unique values in zip code now which was over 800 earlier
```
